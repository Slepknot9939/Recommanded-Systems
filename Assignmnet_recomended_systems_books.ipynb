```python
import pandas as pd
import numpy as np


df=pd.read_csv('book.csv',engine='python')
df.head()
```

```python
data=df.copy()
```

```python
data1=data.drop(['Unnamed: 0'],axis=1)
```

```python
data1.head()
```

```python
data2=data1.rename({'User.ID':'uid','Book.Title':'title','Book.Rating':'rating'},axis=1)
data2.head()
```

```python
len(data2.uid.unique())
```

```python
avrage_range=pd.DataFrame(data2.groupby('title')['rating'].mean())
avrage_range.head()
```

```python
avrage_range['total rating']=pd.DataFrame(data2.groupby('title')['rating'].count())
avrage_range.head()
avrage_range.shape
```

```python
book_user=data2.pivot_table(index='uid',columns='title',values='rating')
book_user.head()
```

```python
book_user.fillna(0, inplace=True)
```

```python
book_user.head()
```

```python
data3=book_user.copy()
data3.head()
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_user.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df.index = data2.uid.unique()
user_sim_df.columns = data2.uid.unique()
```
