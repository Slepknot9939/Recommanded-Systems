```python
import pandas as pd
import numpy as np


df=pd.read_csv('book.csv',engine='python')
df.head()
```

```python
data=df.copy()
```

```python
data.iloc[:,1:]
```

```python
data1=data.rename({'User.ID':'uid','Book.Title':'title','Book.Rating':'ratings'},axis=1)
```

```python
data1.iloc[:,1:]
```

```python
data1=data1.drop('Unnamed: 0',axis=1)
```

```python
data1.head()
```

```python
Average_ratings=pd.DataFrame(data1.groupby('title')['ratings'].mean())
Average_ratings.head()
```

```python
Average_ratings['Total Ratings']=pd.DataFrame(data1.groupby('title')['ratings'].count())
Average_ratings.head()
```

```python
movie_user=data1.pivot_table(index='uid',columns='title',values='ratings')
movie_user.head()
```

```python
correlations=movie_user.corrwith(data1)
correlations.head()
```

```python
recommendation=pd.DataFrame(correlations,columns=['Correlations'])
recommendation.dropna(inplace=True)
recommendation=recommendation.join(Average_ratings['Total Ratings'])
recommendation.head()
```
